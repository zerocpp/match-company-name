```python
import pandas as pd
```

```python
df = pd.read_csv("output/cut-result.csv")
```

```python
df["分词"].values
```

```python
word_tf = dict()
for it in df["分词"].values:
    for i in it.split("|"):
        word_tf[i] = word_tf.get(i, 0) + 1
```

```python
sorted(word_tf.items(), key=lambda it:-it[1])[:10]
```

```python
df = pd.DataFrame(sorted(word_tf.items(), key=lambda it:-it[1]), columns=["词", "词频"])
```

```python
df
```

```python
df.to_csv("output/word-stat.csv", index=0)
```
